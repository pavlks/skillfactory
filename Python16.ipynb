```python
import os
import time
```

```python
def sizeof_fmt(num, suffix="B"):
    for unit in ["", " K", " M", " G"]:
        if abs(num) < 1024.0:
            return f"{num:3.1f}{unit}{suffix}"
        num /= 1024.0
    return f"{num:.1f}Yi{suffix}"
```

```python
ff = dict()
dd = dict()
n = int()
size = int()
```

```python
start = time.time()
for dir, folders, files in os.walk('/home/pavlk/pCloudDrive/'):
    for file in files:
        if file in ff.keys() and sizeof_fmt(os.path.getsize(os.path.join(dir, file))) == ff[file]:
            dd[file] = list([sizeof_fmt(os.path.getsize(os.path.join(dir, file))), dir])
            size += os.path.getsize(os.path.join(dir, file))
        else:
            ff[file] = sizeof_fmt(os.path.getsize(os.path.join(dir, file)))
print(int(time.time() - start), 'seconds')
```
