```python
def currency_name(cur_id):
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'
    response = requests.get(url)
    currencies = response.json()
    for cur in currencies['Valute']:
        if currencies['Valute'][cur]['ID'] == cur_id:
            return currencies['Valute'][cur]['Name']
```

```python
print(currency_name('R01775'))
```

```python
def wiki_header(url):
    resp = requests.get(url)
    page = BeautifulSoup(resp.text, 'html.parser')
    header = page.find('h1')
    return header.text
```

```python
def get_actors(url: str) -> list:
    res = requests.get(url)
    code = BeautifulSoup(res.text, 'html.parser')
    spisok = code.find('ul')
    actors_list = [a.text for a in spisok]
    return actors_list
```

```python
get_actors('https://www.kinopoisk.ru/film/10079/')
```

```python
url = 'https://api.vk.com/method/groups.getMembers'
count = 1000
offset = 0  
user_ids = []
max_count = 100001  
while offset < max_count:  
    print('Выгружаю {} пользователей с offset = {}'.format(count, offset))     
    params = {  
        'group_id': 'vk',  
        'v': 5.95,  
        'count': count,  
        'offset': offset,  
        'access_token': TOKEN  
    }     
    # такой же запрос как в прошлый раз  
    r = requests.get(url, params = params)  
    data = r.json()     
    user_ids += data['response']['items']   

    # увеличиваем смещение на количество строк выгрузки  
    offset += count  
```

```python
token = TOKEN

def get_smm_index(group_name, token):
    url = 'https://api.vk.com/method/groups.getMembers'

    params = {
        'group_id': group_name,
        'v': 5.95,
        'count': 1000,
        'offset': 0,
        'access_token': token
    }
    r = requests.get(url, params = params)
    data = r.json()
    users = data['response']['count']
    
    
    url = 'https://api.vk.com/method/wall.get'
    params = {
        'domain': group_name,
        'filter': 'owner',
        'count': 10,
        'offset': 0,
        'access_token': token,
        'v': 5.95
    }
    response = requests.get(url, params = params)
    r = response.json()
    likes = 0
    comments = 0
    reposts = 0
    for record in r['response']['items']:
        likes += record['likes']['count']
        comments += record['comments']['count']
        reposts += record['reposts']['count']
    total = likes + comments + reposts
    return total / users
```

```python
def get_smm_index(group_name, token):
    url = 'https://api.vk.com/method/wall.get'
    params = {  
    'domain': group_name,  
    'filter': 'owner',  
    'count': 10,  
    'offset': 0,  
    'access_token': token,  
    'v': 5.95  
    }  
    url1 = 'https://api.vk.com/method/groups.getMembers'  
    params1 = {  
    'group_id': group_name,  
    'v': 5.95,  
    'access_token': token  
    }  
    response = requests.get(url, params = params)
    response1 = requests.get(url1, params = params1)
    data = response1.json()['response']['count']
    sum1 = 0
    for record in response.json()['response']['items'][:]:  
        sum1 += (record['comments']['count'] + record['likes']['count'] + record['reposts']['count'])
    smm_index = sum1 / data
    return smm_index
```
